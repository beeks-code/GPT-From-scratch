```python
import torch
```

```python
inputs=torch.rand(size=[2,5])
inputs
```

```python
layer=torch.nn.Sequential(torch.nn.Linear(5,6),torch.nn.ReLU())
```

```python
out=layer(inputs)
```

```python
mean=out.mean(dim=-1,keepdim=True)
var=out.var(dim=-1,keepdim=True)
normalized_out=(out-mean)/torch.sqrt(var)
```

```python
print(f"Mean of Normalized output is {normalized_out.mean(dim=-1,keepdim=True)}")
print(f"Variance of Normalized output is {normalized_out.var(dim=-1,keepdim=True)}")
```

```python
class LayerNormalization(torch.nn.Module):
    def __init__(self,emb_dim):
        super().__init__()
        self.eps=1e-5
        emb_dim=torch.tensor(emb_dim,dtype=torch.float16)
        self.scale=torch.nn.Parameter(emb_dim)
        self.shift=torch.nn.Parameter(emb_dim)
    def forward(self,x):
        self.mean=x.mean(dim=-1,keepdim=True)    
        self.var=x.var(dim=-1,keepdim=True,unbiased=False)
        norm=(x-self.mean)/torch.sqrt(self.var+self.eps)   
        out= norm*self.scale + self.shift 
        return out
    def get_norm_params(self):
        print(f"Means is {self.mean}, \n Variance is {self.var}")
```

```python
test_obj=LayerNormalization(inputs.shape[1])
test_obj(inputs)
test_obj.get_norm_params()
```
